```python
import pandas as pd
import numpy as np

# for graphical visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ax = pd.read_excel("anxity.xlsx")
df_ax = df_ax.rename(columns={'Saverity\nLevel': 'Severity_Level','RelationShip Status':'RelationShip_Status','Current Role':'Current_Role','Family Income':'Family_Income'})
df_ax
```

```python
df_ax['Severity_Level']=df_ax['Severity_Level'].map({'Extremely Severe':4, 'Severe':3,'Moderate':2,'Mild':1,'Normal':0})

df_ax['RelationShip_Status']=df_ax['RelationShip_Status'].map({'Single':0, 'Married':1})

df_ax['Family_Income']=df_ax['Family_Income'].map({'0-20000':0, '20000-50000':1,'50000-100000':2})

df_ax['Current_Role']=df_ax['Current_Role'].map({'Own a Business':2, 'Part Time':1,'Full Time Student':0})

df_ax['Education']=df_ax['Education'].map({'Undergraduate':0})

df_ax['Age']=df_ax['Age'].map({'18-22':0})

df_ax['Gender']=df_ax['Gender'].map({'Female':0,'Male':1})

df_ax

```

```python
df_ax.drop(['id', 'Anxiety','Severity_Level'],axis=1,inplace=True)
df_ax
```

```python
import numpy as np
import pandas as pd

# Assuming df_ax is your DataFrame
mask = (df_ax['Age'] == 0)
random_values = np.random.randint(18, 23, size=mask.sum())
df_ax.loc[mask, 'Age']=random_values
```

```python
df_ax.to_csv('anxity_clean.csv',index=False)
```

```python
df_ax
```

```python
# independent features
X=df_ax.drop("Anxiety1",axis=1)

# dependent feature
y=df_ax["Anxiety1"]
```

```python
X
```

```python

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python

# from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import (r2_score, 
                             mean_absolute_error, 
                             mean_squared_error
                        )


#ML models
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbc = GradientBoostingRegressor(random_state=42, learning_rate=0.1, n_estimators=256, subsample=0.75)
gbc.fit(X_train, y_train)

print('*'*10)

# prediction
train_prediction= gbc.predict(X_train)
test_prediction= gbc.predict(X_test)
print('*'*10)
# evaluation

from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
print('R2 for train data', r2_score(y_train,train_prediction))
print('R2 for test data', r2_score(y_test,test_prediction))
print('*'*10)

print('RMSE for train data', np.sqrt(mean_squared_error(y_train,train_prediction)))
print('RMSE for test data', np.sqrt(mean_squared_error(y_test,test_prediction)))


print('*'*10)
print('*'*10)
print('MSE for train data:',(mean_squared_error(y_train, train_prediction)))
print('MSE for test data:',(mean_squared_error(y_test, test_prediction)))
print('*' * 10)


print('AMSE for train data:',(mean_absolute_error(y_train, train_prediction)))
print('AMSE for test data:',(mean_absolute_error(y_test, test_prediction)))
print('*' * 10)

gb = r2_score(y_test,test_prediction)
```

```python
from xgboost import XGBRegressor
 
Xgb = XGBRegressor(random_state=42)
Xgb.fit(X_train, y_train)

print('*'*10)

# prediction
train_prediction= Xgb.predict(X_train)
test_prediction= Xgb.predict(X_test)
print('*'*10)
# evaluation

from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
print('R2 for train data', r2_score(y_train,train_prediction))
print('R2 for test data', r2_score(y_test,test_prediction))
print('*'*10)

print('RMSE for train data', np.sqrt(mean_squared_error(y_train,train_prediction)))
print('RMSE for test data', np.sqrt(mean_squared_error(y_test,test_prediction)))

print('*'*10)
print('MSE for train data:',(mean_squared_error(y_train, train_prediction)))
print('MSE for test data:',(mean_squared_error(y_test, test_prediction)))
print('*' * 10)


print('AMSE for train data:',(mean_absolute_error(y_train, train_prediction)))
print('AMSE for test data:',(mean_absolute_error(y_test, test_prediction)))
print('*' * 10)

xg = r2_score(y_test,test_prediction)
```
